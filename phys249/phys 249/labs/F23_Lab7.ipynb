**1)** Given the square matrix

- calculate the determinant of A

- calculate the transpose of A

- calculate the inverse of A

You can try this with linalg, but try the first two by hand as well. Note that you can calculate the determinant of a 3x3 matrix 

$$A = \begin{bmatrix}a\ b\ c\\ d\ e\ f \\g\ h\ i\end{bmatrix}$$

with the formula:

$$det(A) = a\cdot det\begin{bmatrix}e\ f \\h\ i\end{bmatrix} - b\cdot det\begin{bmatrix}d\ f \\g\ i\end{bmatrix} + c\cdot det\begin{bmatrix}d\ e \\g\ h\end{bmatrix}$$ 


Det(A) = 1*(2 * 4 - 3 * 2) - 0 * (3 * 4-4 * 2)+1 * (3 * 3 - 2 * 4) = 8 - 6 + 9 - 8 = 3

Trans(A) = $A = \begin{bmatrix}1\ 3\ 4\\ 0\ 2\ 3 \\1\ 2\ 4\end{bmatrix}$

```python
from scipy import linalg
from numpy import array 
a = array([[1,0,1], [3,2,2], [4,3,4]])
res = linalg.inv(a)
print(res)
```
