### The Fibonacci Sequence

Remember that the Fibonacci sequence is calcualted by taking intial terms $F_0 = 1, F_1 = 1$, and then applying the rule $F_{i+1} = F_{i} + F_{i - 1}$. 

1) Caclulate the Fibonnaci sequence $F_i$, for values of $i$ going from 0 to 20, and save the values in a *list*, using the $append$ method to add successive values. Print the resulting list.

```python
fib_list = [1, 1]
for i in range(0, 19, 1):
    fib_list.append(fib_list[i]+fib_list[i+1])

print(fib_list)
```

2) Now repeat this calculation, but save the results in an array (note there are several ways to do this). Print the resulting array.

```python
from numpy import array, zeros

fib_arr = zeros(21, dtype=int)
fib_arr[0] = 1
fib_arr[1] = 1
for i in range(0, 19, 1):
    fib_arr[i+2] = fib_arr[i] + fib_arr[i+1]

print(fib_arr)
```

3) Now take the ratio of successive terms 
$$R_i = {{F_{i+1}}\over{F_{i}}}.$$ 
and save this in an array (you may want to think about how many elements it will have, and what the first/last indices should be). Print the resulting array.


```python
ratios = zeros(20)
for i in range(0, 20, 1):
    ratios[i] = fib_arr[i+1]/fib_arr[i]

print(ratios)
```

4) Use the function *open()* discussed in class to open a file for writing, i.e. with a statement like
$f = open('output.txt','w')$.

Write the values $i, F_i, F_{i+1},$ and $R_i$ to the file, one set per line. Close the file using its $close$ method when you are done.

```python
f = open('output.txt', 'w')
lines = []
for i in range(20):
    # print(i, fib_arr[i], fib_arr[i+1], ratios[i], file=f)
    line = "{} {} {} {}\n".format(i, fib_arr[i], fib_arr[i+1], ratios[i])
    lines.append(line)

f.writelines(lines)
f.close()
```

5) Re-open the file and read the values from it into a (single) array, using the $numpy$ function *loadtxt()*.

```python
from numpy import loadtxt

vals = loadtxt("output.txt")
print(vals)
```

6) Plot $F_i$ versus $i$. (Note `plot a versus z' always means a, the dependent variable is on the vertical axis, while z, the independent variable, is on the horizontal axis.) Make sure to include a $show$ command so your plot displays.

```python
from matplotlib import pyplot as plt

x = vals[:,0]
y = vals[:,1]
print(x, y)

plt.plot(x, y)
plt.show()
```
