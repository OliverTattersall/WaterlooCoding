**a** What is the asteroid's initial position vector in a coordinate system with the same orientation, but with the origin at the location of the Earth? What is its final position vector in this coordinate system?

**b** What are its mean radial and tangential velocity, i.e. the velocity vector $\Delta\vec{P}/\Delta t$ decomposed into radial and tangential components with respect to the Earth? Is there any ambiguity about how to define this?

Answer:

a)
Since (1,0,0) is the new origin, the old origin (0,0,0) now becomes (-1,0,0). Adding that to $P_0$, we get 
$$\vec{P}(t_0) = (-1,1.2,0)$$
Adding that to $P_1$, we get
$$\vec{P}(t_0) = (-1,1.5,0.3)$$

b)
The radial velocity is the change in distance from the origin to the point over change in time. The tangential velocity is the change in angular position over change in time, multiplied by the radius of rotation, however, there is ambiguity of what is the radius of the of the system. For this system we will take the smaller radius as the radius of rotation. There is also ambiguity on whether the coordinate system is in cartesian or cylindrical or spherical. We will assume it is in cartesian.

$$ R_1 = \sqrt{(-1)^2 + (1.5)^2+(0.3)^2} = \sqrt{3.34}$$
$$ \text{Radial velocity } = (R_1 - R_0)/\Delta{t} $$

$$ \text{Radial velocity } = 4.378e10/2592000 = 16890.43 \text{ meters per second in the P(t0) direction} $$
